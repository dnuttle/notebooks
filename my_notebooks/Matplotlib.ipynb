### Line plotting

Line styles:
- \- solid line
- -- dashed line
- : dotted line
- None or space or empty string: no line
Marker (there are many others:
- o Circle marker instead of line
- . point
- , pixel
- v triangle down
- ^ triangle up
- < triangle left
- \> triangle right
- 8 octagon
- s square
- p pentagon
- \* star


```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2]
y = [3, 4]
plt.plot(x, y, label="y")
x2 = [1, 3]
y2 = [3,4]
plt.plot(x2, y2, color='red', linestyle="--", linewidth=2, label="y2")
x3 = [1.5, 4]
y3 = [3, 5]
plt.plot(x3, y3, color='#333aaa', linestyle=":", marker="o", linewidth=2, label="y3")

plt.plot(4, 4, 'o')
#combine color and marker
plt.plot(3.5, 4, 'bo')

plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("A Plot")
plt.legend()
plt.show()
```

### Scatterplot

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 3.9, 4.2, 4.9]
#plot will work but you have to specify a marker and no line
#plot is apparently faster if marker and color don't vary
#plt.plot(x, y, marker="o", linestyle="None")
#Instead
plt.scatter(x, y)
plt.show()
plt.scatter(x, y, color='red', marker="*", s=64)
plt.show()
#Even use latex symbol!
plt.scatter(x, y, marker="$\Delta$", s=200)
plt.show()
```
