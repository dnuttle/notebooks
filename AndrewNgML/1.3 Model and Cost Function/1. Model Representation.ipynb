A training set is fed into an algorithm, and it produces a function denoted $h$ (for hypothesis).

For example, the size of a house might be used to estimate its price.

$h$ maps from $x$ values (independent variable) to $y$ values (dependent variable).

The representation for the $h$ function is:

$$h_{\theta}(x) = \theta_{0} + \theta_{1}x$$

Sometimes $h_{\theta}(x)$ will be written in the shorthand form of $h(x)$.

The goal is to take a dataset of an independent and a dependent variable, and be able to create a line that serves as predictions for new values of $x$.

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8]
y = [10,12,15,13,16,16,18,17]
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y)
plt.plot([1,8], [11, 18])
```

The notation for points in the dataset will be $x^{(i)}, y^{(i)}$, where $i$ is an index referring to a row in the dataset.
